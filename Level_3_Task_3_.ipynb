## Analyze price range vs. online delivery

### Subtask:
Group the data by 'Price range' and 'Has Online delivery' and calculate the proportion of restaurants in each group.


**Reasoning**:
Group the data by 'Price range' and 'Has Online delivery', count the occurrences, unstack, and calculate the proportions for online delivery analysis.



```python
online_delivery_analysis = df.groupby(['Price range', 'Has Online delivery']).size().unstack(fill_value=0)
online_delivery_analysis = online_delivery_analysis.div(online_delivery_analysis.sum(axis=1), axis=0)
display(online_delivery_analysis)
```

## Analyze price range vs. table booking

### Subtask:
Group the data by 'Price range' and 'Has Table booking' and calculate the proportion of restaurants in each group.


**Reasoning**:
Group the data by 'Price range' and 'Has Table booking', calculate the proportion of restaurants in each group, and display the result.



```python
table_booking_analysis = df.groupby(['Price range', 'Has Table booking']).size().unstack(fill_value=0)
table_booking_analysis = table_booking_analysis.div(table_booking_analysis.sum(axis=1), axis=0)
display(table_booking_analysis)
```

## Visualize results

### Subtask:
Create bar plots to visualize the relationship between price range and the availability of online delivery and table booking.


**Reasoning**:
Create bar plots to visualize the relationship between price range and the availability of online delivery and table booking.



```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].bar(online_delivery_analysis.index, online_delivery_analysis['Yes'])
axes[0].set_xlabel('Price Range')
axes[0].set_ylabel('Proportion of Restaurants with Online Delivery')
axes[0].set_title('Online Delivery Availability by Price Range')

axes[1].bar(table_booking_analysis.index, table_booking_analysis['Yes'])
axes[1].set_xlabel('Price Range')
axes[1].set_ylabel('Proportion of Restaurants with Table Booking')
axes[1].set_title('Table Booking Availability by Price Range')

plt.tight_layout()
plt.show()
```
