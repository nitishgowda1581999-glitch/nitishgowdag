## Analyze text reviews

### Subtask:
Preprocess the text reviews (e.g., remove punctuation, convert to lowercase, tokenize) and identify common positive and negative keywords.


**Reasoning**:
The 'Rating text' column will be used as a proxy for sentiment as there is no dedicated text review column. I will map the categorical ratings to simplified positive/negative sentiments.



```python
sentiment_mapping = {
    'Excellent': 'positive',
    'Very Good': 'positive',
    'Good': 'positive',
    'Average': 'negative',
    'Poor': 'negative',
    'Not rated': 'negative'
}

df['Sentiment'] = df['Rating text'].map(sentiment_mapping)

display(df[['Rating text', 'Sentiment']].head())
display(df['Sentiment'].value_counts())
```

**Reasoning**:
I will group the DataFrame by the simplified sentiment categories and describe the distribution of the different rating categories within these groups.



```python
sentiment_distribution = df.groupby('Sentiment')['Rating text'].value_counts().unstack(fill_value=0)

display(sentiment_distribution)
```

```python
import matplotlib.pyplot as plt

sentiment_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Rating Text within Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Rating Text')
plt.tight_layout()
plt.show()
```

## Analyze review length and rating

### Subtask:
Calculate the length of each review, calculate the average review length, and explore the relationship between review length and rating.


**Reasoning**:
Calculate the length of the 'Rating text' column, calculate the average length, and calculate the correlation between 'Review Length' and 'Aggregate rating'. Then, create a scatter plot to visualize the relationship.



```python
df['Review Length'] = df['Rating text'].apply(len)

average_review_length = df['Review Length'].mean()
print(f'Average Review Length: {average_review_length:.2f}')

correlation = df['Review Length'].corr(df['Aggregate rating'])
print(f'Correlation between Review Length and Aggregate Rating: {correlation:.2f}')

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Review Length'], df['Aggregate rating'], alpha=0.5)
plt.title('Review Length vs. Aggregate Rating')
plt.xlabel('Review Length')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()
```

## Analyze Price Range and Services

### Subtask:
Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

**Reasoning**:
Group the data by 'Price range' and calculate the proportion of restaurants offering 'Has Online delivery' and 'Has Table booking' for each price range.

```python
price_service_analysis = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].apply(lambda x: (x == 'Yes').sum() / len(x)).reset_index()
price_service_analysis = price_service_analysis.rename(columns={'Has Online delivery': 'Online Delivery Proportion', 'Has Table booking': 'Table Booking Proportion'})

display(price_service_analysis)
```

### Subtask:
Visualize the relationship between price range and service availability.

**Reasoning**:
Create bar plots to visualize the proportion of restaurants with online delivery and table booking for each price range.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].bar(price_service_analysis['Price range'], price_service_analysis['Online Delivery Proportion'], color='skyblue')
axes[0].set_title('Proportion of Restaurants with Online Delivery by Price Range')
axes[0].set_xlabel('Price Range')
axes[0].set_ylabel('Proportion with Online Delivery')
axes[0].set_xticks(price_service_analysis['Price range'])

axes[1].bar(price_service_analysis['Price range'], price_service_analysis['Table Booking Proportion'], color='lightcoral')
axes[1].set_title('Proportion of Restaurants with Table Booking by Price Range')
axes[1].set_xlabel('Price Range')
axes[1].set_ylabel('Proportion with Table Booking')
axes[1].set_xticks(price_service_analysis['Price range'])

plt.tight_layout()
plt.show()
```
