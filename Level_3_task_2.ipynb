```python
import pandas as pd

df = pd.read_csv('/content/Dataset  (1) (1).csv')
display(df.head())
```

# Task
Analyze the dataset located at "/content/Dataset (1) (1).csv" to identify the restaurants with the highest and lowest number of votes. Additionally, analyze if there is a correlation between the number of votes and the rating of a restaurant. Provide a step-by-step analysis.

## Identify restaurants with the highest votes

### Subtask:
Find the restaurant(s) with the maximum number of votes.


**Reasoning**:
Find the maximum number of votes and filter the DataFrame to display the restaurant(s) with that number of votes.



```python
max_votes = df['Votes'].max()
highest_voted_restaurants = df[df['Votes'] == max_votes]
display(highest_voted_restaurants)
```

## Identify restaurants with the lowest votes

### Subtask:
Find the restaurant(s) with the minimum number of votes.


**Reasoning**:
Find the minimum votes and filter the dataframe to get the restaurants with the minimum votes.



```python
min_votes = df['Votes'].min()
lowest_voted_restaurants = df[df['Votes'] == min_votes]
display(lowest_voted_restaurants)
```

## Analyze correlation between votes and rating

### Subtask:
Calculate the correlation coefficient between the 'Votes' and 'Aggregate rating' columns.


**Reasoning**:
Calculate the Pearson correlation coefficient between 'Votes' and 'Aggregate rating' and store it in a variable, then print the variable.



```python
correlation_votes_rating = df['Votes'].corr(df['Aggregate rating'])
print(correlation_votes_rating)
```

## Visualize correlation

### Subtask:
Create a scatter plot to visualize the relationship between votes and rating.


**Reasoning**:
Create a scatter plot to visualize the relationship between votes and rating as per the instructions.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Votes'], df['Aggregate rating'], alpha=0.5)
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.title('Scatter Plot of Votes vs. Aggregate Rating')
plt.show()
```
